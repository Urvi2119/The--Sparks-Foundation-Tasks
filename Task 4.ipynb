### Cleaning the Data

```python
#First Ten Columns:

ftc = df.iloc[:,:10]
ftc
```

```python
#Applied Drop function on first ten columns:

ftc_up = ftc.drop(["approxdate","extended","resolution"], axis =1 )
ftc_up
```

```python
#Applied Drop function Second Fifteen columns:
stc_up= stc.drop(["multiple"], axis =1 )
stc_up
```

```python
ttc = df.iloc[:,26:41]
ttc
```

```python
ttc_up = ttc.drop(["suicide","attacktype2","attacktype2_txt","attacktype3","attacktype3_txt"], axis =1 )
ttc_up
```

```python
fotc = df.iloc[:,41:56]
fotc
```

```python
fotc_up = fotc.drop(["targtype2","targtype2_txt","targsubtype2","targsubtype2_txt","corp2","target2","natlty2","natlty2_txt","targtype3","targtype3_txt","targsubtype3","targsubtype3_txt","corp3","target3"], axis=1)
fotc_up
```

```python
fitc = df.iloc[:,56:71]
fitc
```

```python
fitc_up = fitc.drop(["natlty3","natlty3_txt","gsubname","gname2","gsubname2","gname3","gsubname3","motive","guncertain1","guncertain2","guncertain3","individual","nperpcap"], axis=1)
fitc_up
```

```python
sitc = df.iloc[:,71:86]
sitc
```

```python
sitc_up = sitc.drop(["weaptype2","compclaim","claimmode3_txt","claimmode3","claim3","claimmode2_txt","claimmode2","claim2","claimmode_txt"], axis=1)
sitc_up
```

```python
setc = df.iloc[:,86:101]
setc
```

```python
setc_up = setc.drop(["nkillter","nkillus","weapsubtype4_txt","weapsubtype4","weaptype4_txt","weaptype4","weapsubtype3_txt","weapsubtype3","weaptype3_txt","weaptype3","weapsubtype2_txt","weapsubtype2","weaptype2_txt"],axis=1)
setc_up
```

```python
eigtc = df.iloc[:,101:116]
eigtc
```

```python
eigtc_up = eigtc.drop(["divert","ndays","nhours","nhostkidus","nhostkid","ishostkid","propextent_txt","nwoundte","nwoundus"],axis=1)
eigtc_up
```

```python
nitc = df.iloc[:,116:135]
nitc
```

```python
nitc_up = nitc.drop(["related","addnotes","nreleased","hostkidoutcome_txt","hostkidoutcome","ransomnote","ransompaidus","ransompaid","ransomamtus"],axis=1)
nitc_up
```

```python
#Updated Dataset after droping all the useless columns:

final=pd.concat([ftc_up,stc_up,ttc_up,fotc_up,fitc_up,sitc_up,setc_up,eigtc_up,nitc_up],axis=1)
final
```

```python
final.columns
```

### Trend of the attacks between 1970-2017

```python
mat.subplots(figsize=(18,5))
sns.countplot(final.iyear,palette='coolwarm',edgecolor=sns.color_palette('dark',7))
mat.xticks(rotation=55)
mat.title('Trend of the terrorist attackes',pad=10.5)
mat.ylabel('Number of attackes')
mat.xlabel('Year')
mat.show()
```

### Top 20 Countries who were hit by high terror attacks

```python
mat.subplots(figsize=(18,5))
sns.barplot(final.country_txt.value_counts()[:20].index,final.country_txt.value_counts()[:20].values,palette='rocket')
mat.xticks(rotation=55)
mat.title('Top 20 countires with high terrorist attacks',pad=10.5)
mat.ylabel('Number of attacks')
mat.xlabel('Country names')
mat.show()
```

### Top 30 safe countires by terror attacks

```python
mat.subplots(figsize=(15,5))
sns.countplot(final.country_txt,order=final.country_txt.value_counts()[185:].index,palette='Greens')
mat.xticks(rotation=50)
mat.title('Safe countries with lowest terror attcks')
mat.ylabel('Number of attacks')
mat.xlabel('Country names')
mat.show()
```

### People killed from 1970-2017 in terrorist attacks

```python
mat.subplots(figsize=(15,5))
sns.barplot(final.groupby('iyear')['nkill'].sum().index,final.groupby('iyear')['nkill'].sum().values,palette='Reds')
mat.xticks(rotation=55)
mat.title('people killed in each year',pad=10.5)
mat.ylabel('Number of people killed in attacks')
mat.xlabel('Year')
mat.show()
```

 ### People killed in number of terrorist attacks

```python
cou1=final.country_txt.value_counts()[:20].to_frame()
gh=final.groupby('country_txt')['nkill'].sum().to_frame()
cou1.columns=['n_attackes']
cou1.merge(gh,left_index=True,right_index=True).plot.bar()
fig=mat.gcf()
fig.set_size_inches(15,5)
mat.xticks(rotation=60)
mat.title('no of attacks v/s people killed')
mat.ylabel('Number of attacks & people killed')
mat.xlabel('country name')
mat.show()
```

### Continental regions with terrorist attacks

```python
values=final.region_txt.value_counts()[:11].values
labelsd=final.region_txt.value_counts()[:11].index
mat.subplots(figsize=(10,8))
mat.pie(values,labels=labelsd,startangle=60)
mat.title('continental wise regions with terrorist attacks')
mat.show()
```

### Most dangerous terrorist groups in the world

```python
mat.subplots(figsize=(10,5))
sns.barplot(final.gname.value_counts()[:15].values,final.gname.value_counts()[:15].index,palette='CMRmap')
mat.title('Most dangerous  terrorist groups in the world')
mat.ylabel('Nmaes of terrorist groups')
mat.xlabel('People killed ')
mat.show()
```

### Success rate of the attacks across the world

```python
mat.subplots(figsize=(5,5))
sns.countplot(final.success,saturation=1)
mat.ylabel('Number of attacks')
mat.show()
```

```python
success=final.success.value_counts()
print(f'The percentage of failuer in a terrorist attack is {np.round((success[0]/success[1])*100,2)}%')
```

### Types of attacking methods used in terrorist attacks

```python
values=final.attacktype1_txt.value_counts().values
labelsd=final.attacktype1_txt.value_counts().index
mat.subplots(figsize=(10,6))
mat.pie(values,startangle=60)
mat.title('Types of attacking methosd used by terrorist')
mat.legend(labelsd,loc='upper right',bbox_to_anchor=(2,0.9))
mat.show()
```

### People killed in various attacking methods

```python
mat.subplots(figsize=(10,6))
labels=final.groupby('attacktype1_txt')['nkill'].sum().sort_values(ascending=False).index
values=final.groupby('attacktype1_txt')['nkill'].sum().sort_values(ascending=False).values
sns.barplot(values,labels)
mat.title('peple killed in various terrorist attacking methods',pad=13)
mat.ylabel('Attacking method')
mat.xlabel('People died')
mat.show()
```

### Key Targets for terrorist attacks

```python
values=final.targtype1_txt.value_counts()[:11].values
labelsd=final.targtype1_txt.value_counts()[:11].index
mat.subplots(figsize=(10,6))
mat.pie(values,labels=labelsd,startangle=60)
mat.title('Targets for terrorist attacks')
mat.show()
```

### Weapon types used in terrorist attacks

```python
values=final['weaptype1_txt'].value_counts().values
labelsd=final['weaptype1_txt'].value_counts().index
mat.subplots(figsize=(10,6))
mat.pie(values,startangle=60)
mat.title('Weapons used in  terrorist attacks')
mat.legend(labelsd,loc='upper right',bbox_to_anchor=(2.3,0.9))
mat.show()
```

### Specific weapons employed in attacked

```python
values=final['weapsubtype1_txt'].value_counts()[:10].values
labelsd=final['weapsubtype1_txt'].value_counts()[:10].index
mat.subplots(figsize=(10,6))
mat.pie(values,startangle=60)
mat.title('Specific weapons used in attacks')
mat.legend(labelsd,loc='upper right',bbox_to_anchor=(1.9,0.9))
mat.show()
```
